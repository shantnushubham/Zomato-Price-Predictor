```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('zomato.csv',encoding='latin-1')
```

```python
data
```

```python
data.shape
```

```python
data.columns
```

```python
data.drop(columns=['Restaurant ID','Is delivering now','Switch to order menu','Rating color','Rating text'],inplace=True)
```

```python
data.columns
```

```python
data=data[data['Country Code']==1]
```

```python
data.columns
```

```python
len(data['City'].value_counts())
```

```python
data=data[data['City'].isin(['New Delhi','Gurgaon','Noida','Faridabad'])]
```

```python
data.shape
```

```python
data.columns
```

```python
data.drop(columns=['Country Code','Restaurant Name','Address','Locality','Locality Verbose','Currency'],inplace=True)
```

```python
data.columns
```

```python
data['Has Table booking'].replace({'Yes':1,'No':0},inplace=True)
```

```python
data['Has Table booking'].value_counts()
```

```python
data.shape
```

```python
data['Has Online delivery'].replace({'Yes':1,'No':0},inplace=True)
```

```python
data['Has Online delivery'].value_counts()
```

```python
data.shape
```

```python
data.columns
```

```python
data.corr()['Average Cost for two']
```

```python
data.drop(columns=['Has Online delivery','Price range'],inplace=True)
```

```python
data.shape
```

```python
data.columns
```

```python
data.info()
```

```python
a=data.groupby('Cuisines')['Average Cost for two'].mean().sort_values(ascending=False).reset_index()
```

```python
data=data.merge(a, on="Cuisines")
```

```python
data.drop(columns=['Cuisines'],inplace=True)
```

```python
data.head()
```

```python
data.rename(columns={"Average Cost for two_x":"cost", "Average Cost for two_y":"cuisine"},inplace=True)
```

```python
y=data.iloc[:,3].values
```

```python
y
```

```python
y.shape
```

```python
X=data.drop(columns=['cost']).iloc[:,:].values
```

```python
X
```

```python
X.shape
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
X[:,0]=encoder.fit_transform(X[:,0])
```

```python
X[:,0]
```

```python
X.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
onencoder=OneHotEncoder(categorical_features=[0])
```

```python
X=onencoder.fit_transform(X).toarray()
```

```python
X.shape
```

```python
X[:,0:4]
```

```python
X=X[:,1:]
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
from sklearn.linear_model import LinearRegression
regressor1= LinearRegression()
```

```python
regressor1.fit(X_train, y_train)
```

```python
y_pred=regressor1.predict(X_test)
```

```python
y_test.shape
```

```python
y_pred.shape
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor2=DecisionTreeRegressor()
```

```python
regressor2.fit(X_train,y_train)
```

```python
y_pred2=regressor2.predict(X_test)
```

```python
r2_score(y_test,y_pred2)
```

```python
regressor1.intercept_
```

```python
regressor1.coef_
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor3=RandomForestRegressor(n_estimators=300)
```

```python
regressor3.fit(X_train, y_train)
```

```python
y_pred3=regressor3.predict(X_test)
```

```python
r2_score(y_test,y_pred3)
```

```python
from sklearn.svm import SVR
regressor4=SVR()
```

```python
regressor4.fit(X_train,y_train)
```

```python
y_pred4=regressor4.predict(X_test)
```

```python
r2_score(y_test,y_pred4)
```
